### ERA5 Boxplot

```python
# Define the figure
plt.figure(figsize=(16, 16))

means = e5df[mean_cols]


# Make the plot
sns.boxplot(data=means, palette="Set2")
plt.title("Boxplots of Mean PWV Values (ERA5)", fontsize=32)
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Stations", fontsize=22)
plt.ylabel("Values", fontsize=22)
plt.show()
```

### ERA5 Heatmap

```python
# get the corellation matrix
correlation_matrix = means.corr()
print(correlation_matrix)

# plot the matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="RdYlGn", cbar=True) # cbar is the color bar(legend), annot is the annontation
plt.title("ERA5 Heatmap of Mean PWV Values", fontsize=22)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
