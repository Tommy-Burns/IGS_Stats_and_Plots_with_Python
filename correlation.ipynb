```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import seaborn as sns
import numpy as np
import os
```

```python
data_path = os.path.join(os.getcwd(), 'data')
# print('Data path:', data_path)
era5_data = os.path.join(data_path, 'STATs_era5-Tm.csv')
gtrop_data = os.path.join(data_path, 'STATs_Gtrop-Tm.csv')

e5df = pd.read_csv(era5_data)
gtdf = pd.read_csv(gtrop_data)
```

```python
corr_cols = [item for item in e5df.columns if "R_" in item][:6]
corr_tc_cols = [item for item in e5df.columns if "R_" in item][6:]

print(corr_cols)
print(corr_tc_cols)
```

### ERA5 Corellation Map

```python
correlation_columns = e5df[corr_cols]
correlation_matrix = correlation_columns.corr()

# Plot the matrix
fig, ax = plt.subplots(figsize=(12, 12))
cax = ax.matshow(correlation_matrix, cmap="RdYlGn")

# Add annotation
for i in range(correlation_matrix.shape[0]):
    for j in range(correlation_matrix.shape[1]):
        ax.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}', ha='center', va='center', color='black', fontsize=12)

plt.colorbar(cax, shrink=0.8)
plt.title("Correlation (R)", fontsize=22)

# Set axis labels and move ticks to the bottom
labels = correlation_columns.columns
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels, fontsize=13, rotation=65)
ax.set_yticklabels(labels, fontsize=13)

# Remove white grid lines
ax.tick_params(axis='both', which='both', length=0)
ax.xaxis.set_ticks_position('bottom')  # Move x-axis ticks to the bottom
ax.yaxis.set_ticks_position('left')    # Move y-axis ticks to the left

ax.grid(visible=False)

# plt.show()
plt.savefig(r".\correlation\ERA5 Correlation.jpeg", dpi=150)
```
